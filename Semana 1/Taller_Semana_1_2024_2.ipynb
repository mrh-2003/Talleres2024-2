```python
#Incluir librerias
import math as m
#Datos de entrada
arista = float(input("Ingrese la arista: "))
#Operaciones
area =  m.pow(arista, 2) * m.sqrt(3)
volumen = m.sqrt(2) / 12 * m.pow(arista, 3)
#Datos de salida
print(f"El volumen es: {volumen} m3")
print(f"El area es: {area} m2")
```

```python
metros_al = int(input("Ingrese los metros de alambre: "))
rollo_500 = metros_al // 500
metros_al = metros_al % 500
rollo_300 = metros_al // 300
metros_al = metros_al % 300
rollo_75 = metros_al // 75
residuo = metros_al % 75
print("Cantidad de rollos de 500:", rollo_500)
print("Cantidad de rollos de 300:", rollo_300)
print("Cantidad de rollos de 75:", rollo_75)
print("El ultimo rollo tendra:", residuo)
```

```python
cantidad = int(input("Ingrese cantidad: "))
codigo = int(input("Ingrese codigo producto: "))

if codigo == 1:
  precio = 15.75
elif codigo ==2:
  precio = 21
elif codigo == 3:
  precio = 8.5
elif codigo == 4:
  precio = 25
else:
  precio = 13.25
monto_pagar = cantidad * precio

print("El monto a pagar es:", monto_pagar)
```

```python
salario = float(input("Ingrese el salario mensual: "))
cantidad =  int(input("Ingrese número de personas a cargo: "))

if cantidad == 1:
  if salario <= 500:
    tipo_linea = 'P'
  else:
    tipo_linea = 'O'
elif cantidad >=2 and cantidad <= 4:
  if salario <= 750:
    tipo_linea = 'P'
  else:
    tipo_linea = 'O'
else:
  if salario <= 1000:
    tipo_linea = 'P'
  else:
    tipo_linea = 'O'


print("Tipo de línea al que puede acceder:", tipo_linea)
```
