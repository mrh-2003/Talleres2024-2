```python
import numpy as np
matriz = np.ones((7, 7))
matriz[:,:3] = 0
matriz[:,3] = 0.5
matriz[-1:,3] = 0.7

prom = np.mean(matriz[-1])

print("El promedio de la ultima fila es: ", prom)
print("La matriz es: ", matriz)
```

```python
import numpy as np
matriz = np.random.rand(5, 5)
print("La matriz es: ", matriz)
i = 0
for fila in matriz:
  j = 0
  for col in fila:
    if col > 0.5:
      print(f"La posicion es fila={i}, columna={j} y el valor es {col}")
    j+=1
  i+=1
```

```python
import numpy as np

matriz = np.array([[-44, 12], [12.0, 51], [1300, -5.0]])
print("Matriz Original:", matriz)
matriz[1] -= 5
print("Matriz Menos 5:", matriz)
matriz *= 2
print("Matriz Por 2:", matriz)
matriz[:2] /= -5
print("Dividir Por -5:", matriz)
print("Imprimir ultima fila:", matriz[-1])
print("Imprimir suma matriz:", np.sum(matriz))
print("Imprimir promedio:", np.mean(matriz[:2]))
```
