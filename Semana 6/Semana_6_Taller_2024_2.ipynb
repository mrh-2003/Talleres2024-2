```python
from datetime import datetime, timedelta

def dias_desde_mi_nacimiento(fecha_nacimiento):
  fecha_hoy = datetime.now()
  dias_transcurridos = (fecha_hoy  - fecha_nacimiento).days
  dias = ['Domingo', 'Lunes', 'Martes', 'Miercoles', 'Jueves', 'Viernes', 'Sabado']
  dia_nacimiento = dias[int(fecha_nacimiento.strftime('%w'))]
  cumpleaños_actual = datetime(year= fecha_hoy.year, month= fecha_nacimiento.month, day= fecha_nacimiento.day)
  if cumpleaños_actual < fecha_hoy:
    cumpleaños_actual = cumpleaños_actual.replace(year= datetime.now().year + 1)
  dias_transcurridos_prox_cumple = abs((fecha_hoy  - cumpleaños_actual).days)
  print("La cantidad de dias transcurridos es:", dias_transcurridos)
  print("La cantidad de dias transcurridos para mi proximo cumpleaños es:", dias_transcurridos_prox_cumple)
  print("El dia de mi nacimiento fue:", dia_nacimiento)

while True:
  try:
    fecha = input("Ingrese la fecha en el formato AAAA/MM/DD: ")
    fecha_nacimiento = datetime.strptime(fecha, "%Y/%m/%d")
    break;
  except:
    print("Ingresa bien la fecha, no seas gil")
dias_desde_mi_nacimiento(fecha_nacimiento)
```

```python
while True:
  try:
    fecha = input("Ingrese la fecha de prestamo AAAA/MM/DD: ")
    fecha_prestamo = datetime.strptime(fecha, "%Y/%m/%d")
    break;
  except:
    print("Ingresa bien la fecha, no seas gil")

while True:
  try:
    fecha = input("Ingrese la fecha de pago AAAA/MM/DD: ")
    fecha_pago = datetime.strptime(fecha, "%Y/%m/%d")
    break;
  except:
    print("Ingresa bien la fecha, no seas gil")

tasa = -1
while not(tasa > 0 and tasa <= 1):
  try:
    tasa = float(input("Ingrese la tasa: "))
  except ValueError:
    print("La tasa va del 0 al 1")

capital = -1
while not(capital > 0 ):
  try:
    capital = float(input("Ingrese el capital: "))
  except ValueError:
    print("El capital es un numero")

dias = (fecha_pago - fecha_prestamo).days

intereses = capital * ((1 + tasa)**(dias/360) - 1)

print("Los intereses a pagar son: ", intereses)
```

```python
while True:
  try:
    fecha = input("Ingrese la fecha de la primera cita AAAA/MM/DD: ")
    fecha_cita = datetime.strptime(fecha, "%Y/%m/%d")
    break;
  except:
    print("Ingresa bien la fecha, no seas gil")

if fecha_cita.day < 20:
  mes = fecha_cita.month
else:
  mes = fecha_cita.month + 1

anio = fecha_cita.year

for i in range(8):

  if mes == 13:
    mes = 1
    anio = anio + 1

  cita = datetime(year = anio, month = mes, day = 20)

  if int(cita.strftime('%w')) == 0:
    cita = cita.replace(day=21)

  mes += 1

  print(f"La cita numero {i + 1} es {cita}")
```
