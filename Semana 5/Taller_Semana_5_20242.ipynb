```python
import random as r
def generar_valores(p_min, p_max):
  lista = []
  while len(lista) < 5:
    lista.append(r.randint(p_min, p_max))
    lista = list(set(lista))
  return lista

def crear_bingo():
  bingo = {
      'B':  generar_valores(1, 15),
      'I':  generar_valores(16, 30),
      'N':  generar_valores(31, 45),
      'G':  generar_valores(46, 60),
      'O':  generar_valores(61, 75),
  }
  return bingo
def mostrar_bingo(bingo):
  for titulo in bingo.keys():
    print(titulo, end="\t")
  print()
  for i in range(5):
    for titulo in bingo.keys():
      print(bingo[titulo][i], end='\t')
    print()

bingo = crear_bingo()
mostrar_bingo(bingo)
```

```python
class CuentaBancaria:
  def __init__(self, nombre, cantidad, tasa_interes):
    self.__nombre = nombre
    self.__cantidad = cantidad
    self.__tasa_interes = tasa_interes
  def aplica_interes(self):
    self.__cantidad = self.__cantidad + self.__cantidad * self.__tasa_interes / 100
  def __str__(self):
    print(f"Nombre:\t{self.__nombre}\nCantidad:\t{self.__cantidad}\nTasa Interes:\t{self.__tasa_interes}")
  def get_nombre(self):
    return self.__nombre
  def set_interes(self, interes):
    if interes > 0:
      self.__tasa_interes = interes

obj = CuentaBancaria('Juan de Aroma', 1000, 3)
obj.__str__()
obj.aplica_interes()
obj.__str__()
obj.set_interes(2)
obj.__str__()
obj.aplica_interes()
obj.__str__()
```

```python
class Articulo:
  def __init__(self, nombre,precio):
    self.__nombre = nombre
    self.__precio = precio
  def __str__(self):
    return f"El nombre del articulo es {self.__nombre} y su precio es {self.__precio : 0.2f}"

obj_articulo = Articulo("Galleta", 15.63255)
obj_articulo.__str__()
```

```python
class Ticket:
  def __init__(self, precio, hora):
    self.__precio = precio
    self.__hora = hora
  def __str__(self):
    return f"Ticket({self.__precio}, {self.__hora})"
  def es_noche_hora(self):
    hh = int(self.__hora[:2])
    return hh >= 18
  def total_descuento(self, n):
    if n >=5 and n <=9:
      return 10
    elif n >= 10:
      return 20
    else:
      return 0

tk = Ticket(154, '18:50')
print(tk.__str__())
if tk.es_noche_hora():
  print("Es noche")
else:
  print("No es noche")
print("El porcentaje de descuento es: ", tk.total_descuento(15))
```
